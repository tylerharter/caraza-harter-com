```python
# Example 1a:  let's pick a number to classify English words as either short or long
# A word in English is "short" if it has _____ letters or less
# otherwise it is "long"
# talk to your neighbor and determine a number

word = input("enter a word ")
word_length = len(word)

if word_length <= 6:  # conditional execution
    print(word, "is short")
    print("short words are ........")

if word_length > 6:
    print(word, "is long")
    print("long words are ")
    
# then...add an extra print statement inside each if block
```

```python
# Example 1b:  Let's rewrite the code using an else statement
# this is called 'alterate execution'

word = input("enter a word ")
word_length = len(word)

if word_length <= 6:  # conditional execution
    print(word, "is short")
    print("short words are fun")
else:
    print(word, "is long")
    print("long words are interesting")

```

```python
# Example 2: Let's classify people who are less than 18 years of age
# what classifications would you use?
# a person who is _____ years old is a _________
# and so on.....

# infant/baby 0-1
# toddler 2-3
# child/kid 4-12
# teen 13-17
# adult 18 and over

age = int(input("enter your age "))
print ("you are a ", end="")

if 0 <= age <= 1:
    print("infant")
if 2 <= age <= 3:
    print("toddler")
if 4 <= age <= 12:
    print("kid")
if 13 <= age <= 17:
    print("teen")
if age >= 18:
    print("adult")
```

```python

#let's redo Example 2 with chained chained conditions
age = int(input("enter your age"))
print ("you are a ", end="")


if age <= 1:
    print("infant")
elif age <= 3:
    print("toddler")
elif age <= 12:
    print("kid")
elif age <= 17:
    print("teen")
else:
    print("adult")
```

```python
# add a sanity check if statement to not allow ages <0 or >18
age = int(input("enter an age between 0 and 18 "))
if age < 0 or age > 18:
    print ("this is not for you ")
else:
    print ("you are a ", end="")
    if age <= 1:
        print("infant")
    elif age <= 3:
        print("toddler")
    elif age <= 12:
        print("kid")
    elif age <= 17:
        print("teen")
    else:
        print("adult")
```

```python
import math

def piecewise(x):
    '''defines the piecwise function shown above'''
    if x <= -2:
        return 2
    elif x <= 0:
        return abs(x)
    else:
        return math.floor(x)

piecewise(1.5)
 
```
