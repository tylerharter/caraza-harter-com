# Recursion Demo 2: Recursive List Search

```python
def contains(n, numbers):
    for i in numbers:
        if type(i) == list:
            if contains(n, i):
                return True
        else:
            if n == i:
                return True
    return False

data = [1,2,[4,[[3],[8,9]],5,6]]
print(data) # Replace this with pretty_print (recursive) function call :)
for i in range(13):
    print(i, contains(i, data))
```
