```python
#DEMO 3: State Populations

prefixURL = "https://www.msyamkumar.com/cs220/s20/materials/lectureDemo_code/lec-31/data/"
r = requests.get(prefixURL + "state_files.txt")
r.raise_for_status()
stateFiles = r.text.split("\n")
stateFiles
```

```python
rows = []

for state in stateFiles:
    url = prefixURL + state
    r = requests.get(url)
    r.raise_for_status()
    data = r.json()
    data["name"] = state
    rows.append(data)

df = DataFrame(rows)
df
```
