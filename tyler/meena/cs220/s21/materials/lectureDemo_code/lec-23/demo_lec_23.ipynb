# Recursion Demo 1: Pretty Print

## What is the simplest example for this problem? This determines the base case

```python
# Stand-alone list passed as argument to pretty_print function.
```

## How can you define self-reference here?

```python
# Self-reference can be defined by passing nested list as argument to pretty_print function.
```

## Let's write the code for the recursive function.

```python
def pretty_print(items):
    for v in items:
        if type(v) == list:
            #Recursive case
            pretty_print(v)
        else:
            #Base case
            print("* " + v)
    
data = ["A", ["1", "2", "3",], "B", ["4", ["i", "ii"]]]
#data = ["A", "B", "C"]
pretty_print(data)
```

## How can we handle indent for nesting here?
- I am leaving the above cell, for your learning purpose
    - It is not recommended to have two function definitions for the same function. 
    - You should give preference to updating the function within the previous cell

```python
#Defining a new parameter called "indent" to take care of nesting level.
```

```python
def pretty_print(items, indent = 0):
    for v in items:
        if type(v) == list:
            #Recursive case
            pretty_print(v, indent + 1)
        else:
            #Base case
            spaces = " " * indent
            print(spaces + "* " + str(v))
    
data = ["A", ["1", "2", "3",], "B", ["4", ["i", "ii"]]]
#data = ["A", "B", "C"]
pretty_print(data)
```

# Recursion Demo 2: Recursive List Search

```python
def contains(n, numbers):
    for i in numbers:
        if type(i) == list:
            if contains(n, i):
                return True
        else:
            if n == i:
                return True
    return False

data = [1,2,[4,[[3],[8,9]],5,6]]
print(data) # Replace this with pretty_print (recursive) function call! How can you fix 
            #pretty_print to incorporate this?
pretty_print(data)
for i in range(13):
    print(i, contains(i, data))
```
