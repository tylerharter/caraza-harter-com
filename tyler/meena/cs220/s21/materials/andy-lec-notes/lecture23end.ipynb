```python
# example 2...the Collatz Conjecture problem 
# https://en.wikipedia.org/wiki/Collatz_conjecture
# run this in Python Tutor on your ownn

def collatz(n):
    print("in collatz with n = " ,  n)
    if n == 1:
        return 1 # base case
    elif n % 2 == 0:
        return collatz(n//2)
    else:
        return collatz (3*n+1)

collatz(13)   # try other numbers
```

```python
# Write a recursive function to search *ANY* list of lists/tuples for a given word

def search_recursive(target, directory):
    '''searches for target in directory'''
    print("entering function with " , directory)
    for thing in directory:
        #print("checking", thing, type(thing))
        if (type(thing) == str) and (thing == target):
            return True
        elif (type(thing) == list) or (type(thing) == tuple):
            if search_recursive(target, thing) == True:
                return True
            else: # not found, keep searching
                continue # check the next thing in directory
    return False # not found in directory

search_recursive("pear", fav_stuff)
```

```python
# write a function that prints with indenting
# 
def print_with_indenting(directory, indent):
    for thing in directory:
        if type(thing) == list or type(thing) == tuple:
            print("   " * indent + str(type(thing)))
            print_with_indenting(thing, indent + 1)
        else:
            print("   " * indent + str(thing))
        
print_with_indenting(fav_stuff, 0)
```
