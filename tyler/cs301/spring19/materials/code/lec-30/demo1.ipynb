```python
import requests
import pandas as pd
```

```python
base_url = 'https://tyler.caraza-harter.com/cs301/fall18/materials/code/lec-28/'

# lookup list of state files
r = requests.get(base_url + 'state_files.txt')
r
```

```python
assert(r.status_code) == 200
```

```python
r.text
```

```python
# get list of state files
state_files = r.text.split('\n')
state_files
```

```python
# let's fetch each state file and parse the JSON it contains

# key: state_file name
# val: dictionary of population stats by year
state_data = {}

for name in state_files:
    url = base_url + name
    r = requests.get(url)
    r.raise_for_status()
    data = r.json()
    print(name, data)
    state_data[name] = data
```

```python
state_data
```

```python
# let's create a pandas table from a dictionary of lists.
# we'll have a row for each state and a column for each year.
# each cell will contain a population
table = {'state':[], '2000':[], '2010':[], '2015':[]}

for filename in state_data:
    state_name = filename.split('.')[0].replace('_', ' ')
    row = state_data[filename]
    
    table['state'].append(state_name)
    for year in ('2000', '2010', '2015'):
        table[year].append(row[year])
table
```

```python
df = pd.DataFrame(table)
df
```
