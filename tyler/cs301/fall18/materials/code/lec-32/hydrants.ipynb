```python
import pandas as pd
import sqlite3
```

```python
c = sqlite3.connect("hydrants.db")
```

```python
pd.read_csv("Fire_Hydrants.csv").to_sql("hydrants", c, replace=True, index=False)
```

```python
df = pd.read_sql("select year_manufactured as year, nozzle_color as color, style, count(*) as count from hydrants group by year_manufactured, nozzle_color, style order by count desc", c)
df.dropna()
```

```python
pd.read_sql("select * from hydrants", c)
```
