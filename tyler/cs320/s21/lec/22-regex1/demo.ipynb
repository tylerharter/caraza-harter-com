## Regular Expressions

```python
import re

# from DS100 book...
def reg(regex, text):
    """
    Prints the string with the regex match highlighted.
    """
    print(re.sub(f'({regex})', r'\033[1;30;43m\1\033[m', text))
```

```python
# character class
reg("[aA]", s1)
```

```python
reg(r"[aeiouAEIOU]", s1) # all the vowels
```

```python
reg(r"[^aeiouAEIOU]", s1) # all the non vowels
```

```python
# character range
reg(r"[A-Z]", s1)
```

```python
reg(r"[\.\-,]", s1)
```

```python
reg(r"\d", s1) # meta character (digit in this case)
```

```python
reg(r"\w", s1) # word (tabs, spaces, newlines)
```

```python
reg(r"\s", s1) # whitespace (tabs, spaces, newlines)
```

```python
reg(r"\S", s1) # NOT whitespace (tabs, spaces, newlines)
```

```python
reg(r".", s1) # match everything except newline
```

```python
reg(r"\.", s1) # match everything except newline
```

# Repetition

```python
reg(r"w{3}", s1)
```

```python
reg(r"w{2}", s1) # exact number
```

```python
# match everything in parens

# bad attempt 1
reg(r"(.*)", s1)
```

```python
# bad attempt 2
reg(r"\(.*\)", s1) # * is "greedy"
```

```python
reg(r"\(.*?\)", s1) # *? is "not greedy"
```

```python
# ?    0 or 1
# *    any number, preferably more
# *?   any number, preferably fewer
# +    at least one, preferably more
# +?   at least one, preferably fewer
```

```python
# find the first sentence.

# bad attempt:
reg(r"[^\.]+\.", s1)
```

```python
# ^  not -- flip the char class (inside of brackets)
# ^  anchor to beginning of str (outside of brackets)
# $  anchor to end of the string
reg(r"^[^\.]+?\.", s1)
```

```python
# want the first 2 sentences
reg(r"^([^\.]+?\.){2}", s1)
```

```python
# find the last word
reg(r"[^\s]+$", s1)
```

```python
# well, technically it's finding a valid phone number
# but this probably isn't what you imagined.
# I won't cover it in 320, but you would need a
# "negative lookbehind assertion" to make sure the
# preceding text wasn't digits or dashes:
# https://docs.python.org/2/library/re.html
reg(r"((\d-)?\d{3}-)?\d{3}-\d{4}", "1-123-4567")
```
